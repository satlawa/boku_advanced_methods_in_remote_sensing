## Spectral filters

First, we have to calculate the diference Fourier Transform (dft function)

```python
dft = cv2.dft(np.float32(I), flags=cv2.DFT_COMPLEX_OUTPUT)
# shift the zero-frequncy component to the center of the spectrum
dft_shift = np.fft.fftshift(dft)
# Calculate magnitude (i.e. square root of the sum of the squares of the real and imaginary parts)
magnitude = cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1])
# save image of the image in the fourier domain: logarithmic streching to not lost visual detail
magnitude_fd = 20*np.log(magnitude)

plt.figure(), plt.imshow(magnitude_fd, cmap='gray')
plt.title('Frequency domain')
plt.axis('off')
plt.show()
```

Second, we should define the filter. In this case is low pass frequency filter. The size of the filter is 50.

```python
size_fil = 50
rows, cols = I.shape
crow, ccol = rows//2, cols//2
# create a mask first, center square is 1, remaining all zeros
mask = np.zeros((rows, cols,2), np.uint8)
mask[crow-size_fil:crow+size_fil, ccol-size_fil:ccol+size_fil] = 1
plt.figure(), plt.imshow(mask[:,:,0], cmap='gray')
plt.title('Low pass mask')
plt.axis('off')
plt.show()
```

We have to multiply the frequency image by the filter mask, then shift the zero-frequency to the origin (ifftshift function) and apply the invers of discrete Fourier transform (idft function).

```python
# apply mask and inverse DFT
fshift = dft_shift*mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:, :, 0], img_back[:, :, 1])

plt.figure(), 
plt.imshow(img_back, cmap='gray')
plt.title('Low Pass Filter')
plt.axis('off')
plt.show()
```

If we would like a high pass frequency filter, we have to define in the opposite way than the low pass. In this case the size of the filter is 3.

```python
size_fil = 3
mask = np.ones((rows, cols, 2), np.uint8)
mask[crow-size_fil:crow+size_fil, ccol-size_fil:ccol+size_fil] = 0
plt.figure(), plt.imshow(mask[:,:,0], cmap='gray')
plt.title('High Pass Mask')
plt.axis('off')
plt.show()
```

The way to apply and the transformations here are equal than low pass filter.

```python
# apply mask and inverse DFT
fshift = dft_shift*mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:, :, 0], img_back[:, :, 1])

plt.figure(), plt.imshow(img_back, cmap='gray', vmax=np.percentile(img_back, 99))
plt.title('High Pass Filter')
plt.axis('off')
plt.show()
```

## Textures

```python
d = [1] # distance
theta = [0] #[0, np.pi/4, np.pi/2, 3*np.pi/4] DIRECTION
props = ('contrast', 'homogeneity','dissimilarity', 'ASM','energy','correlation') # statistic properties to extract
levels = 256 # levels of grey of the image
win = 19 # moving window
feature_map = haralick_textures(I_U8, win, d, theta, levels, props)
# feature_map is a cube of matrices with each statistic property is stack in the third dimension. So, the size of feature_map is 400x400x6
```
